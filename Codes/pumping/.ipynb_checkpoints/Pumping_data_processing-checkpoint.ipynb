# Arizona

```python
import os
import numpy as np
import pandas as pd
from glob import glob
import geopandas as gpd
```

```python
# collecting all csvs
Arizona_pumping_csvs = glob(os.path.join('../../Data_main/Pumping/Arizona/raw', '*.csv'))
Arizona_pumping_csvs
```

```python
# compiling all csvs into one dataframe
compiled_pump_df = pd.DataFrame()

for csv in Arizona_pumping_csvs:
    pump_df = pd.read_csv(csv)
    compiled_pump_df = pd.concat([compiled_pump_df, pump_df])

compiled_pump_df.head()
```

```python
compiled_pump_df.columns
```

```python
# converting corodinates
compiled_pump_df['Lat_NAD83'] = compiled_pump_df['LAT_DEGREES'] + (compiled_pump_df['LAT_MINUTES']/60) + (compiled_pump_df['LAT_SECONDS']/3600)
compiled_pump_df['Lon_NAD83'] = -(compiled_pump_df['LONG_DEGREES'] + (compiled_pump_df['LONG_MINUTES']/60) + (compiled_pump_df['LONG_SECONDS']/3600))

# selecting needed columns
compiled_pump_df = compiled_pump_df[['Well Id', 'AMA INA', 'Basin', 'YEAR',
       'Right Type', 'AF Pumped', 'Water Type', 'Lat_NAD83', 'Lon_NAD83']]

# Right types
print('Right Types:', '\n', compiled_pump_df['Right Type'].unique())

# Water Type
print('Water Types:', '\n', compiled_pump_df['Water Type'].unique())
```

-----

`Relevant Right Type: 'IRRIGATION USE', 'IRRIGATION DISTRICT (GW ONLY)'`

`Relevant Water Type: 'GROUNDWATER'`

```python
# filtering for correct 'Right Type' and 'Water Type'
compiled_pump_df = compiled_pump_df[(compiled_pump_df['Right Type'].isin(['IRRIGATION USE', 'IRRIGATION DISTRICT (GW ONLY)'])) & 
                                (compiled_pump_df['Water Type'] == 'GROUNDWATER')]
compiled_pump_df
```

## v1

### Getting well lat/lon infor from GWSI database

The original pumping database has some missing coordinates. Combining with GWSI databse can fill some of that. Still, some of the gaps remain. Here, we will create two versions of the annual basin-wise pumping summary file. The `v1` version compiles only the wells that has location info. The `v2` version compiles all pumping estimates annualy in a basin irrespective of location availability. We will use `v2` for comparing our model generated estimates with in-situ annual pumping (aggregated for a basin), but still making `v1` to keep track of things (will further work on `v1` to get get all well location info somehow). 

__Note that__ `v1` has individual well info that has location info while `v2` compiles all well data annualy for a basin even if there are missing well lcoation info; therefore, the total AF pumped for a basin in a particular year will mismatch from these two versions. For annual comparison `v2` is accurate.

```python
# keeping a copy of the compiled pumping databases to use in `v2`
original_compiled_df = compiled_pump_df.copy()
```

```python
# The primary well database has missing well coordinates. Retrieving that from GWSI database
GWSI_df = pd.read_csv('../../Data_main\Pumping/Arizona/raw/GWSI_sites.csv')
GWSI_df.columns
```

```python
# selecting only required columns
GWSI_df = GWSI_df[['REG_ID', 'DD_LAT', 'DD_LONG']]
GWSI_df = GWSI_df.dropna()
GWSI_df
```

```python
# merging pumping dataframe and GWSI records on 'Well Id' and 'REG_ID'
compiled_pump_df = compiled_pump_df.merge(GWSI_df, left_on='Well Id', right_on= 'REG_ID')
compiled_pump_df.columns
```

```python
# renaming some columns
compiled_pump_df = compiled_pump_df.rename(columns={'YEAR': 'year', 'AF Pumped': 'AF_pumped', 
                                                    'Right Type': 'Right_type', 'Water Type': 'Water _type'})
compiled_pump_df.head()
```

```python
# Filling missing lat/long values
compiled_pump_df['Lat_NAD83'] = np.where(np.isnan(compiled_pump_df['Lat_NAD83']) & (~np.isnan(compiled_pump_df['DD_LAT'])), 
                                                                                    compiled_pump_df['DD_LAT'], np.nan)
compiled_pump_df['Lon_NAD83'] = np.where(np.isnan(compiled_pump_df['Lon_NAD83']) & (~np.isnan(compiled_pump_df['DD_LONG'])), 
                                                                                    compiled_pump_df['DD_LONG'], np.nan)

# Dropping nan values
compiled_pump_df = compiled_pump_df.dropna()

# Drop columns
compiled_pump_df = compiled_pump_df.drop(columns=['REG_ID', 'DD_LAT', 'DD_LONG'])

# Dropping duplicates rows if any
compiled_pump_df = compiled_pump_df.drop_duplicates(subset=['Well Id', 'AMA INA', 'year', 'Lat_NAD83', 'Lon_NAD83'])

compiled_pump_df
```

```python
# saving as csv
compiled_pump_df.to_csv('../../Data_main/Pumping/Arizona/pumping_AZ_v1.csv', index=False)
```

## v2
### Summing annual pumping values for AMA/INA

The original pumping database has some missing coordinates. Combining with GWSI databse can fill some of that. Still, some of the gaps remain. Here, we will create two versions of the annual basin-wise pumping summary file. The `v1` version compiles only the wells that has location info. The `v2` version compiles all pumping estimates in a basin irrespective of location availability. We will use `v2` for comparing our model generated estimates with in-situ annual pumping (aggregated for a basin), but still making `v1` to keep track of things (will further work on `v1` to get get all well location info somehow).

__Note that__ `v1` has individual well info that has location info while `v2` compiles all well data annualy for a basin even if there are missing well lcoation info; therefore, the total AF pumped for a basin in a particular year will mismatch from these two versions. For annual comparison `v2` is accurate.

```python
# summping pumping values for AZ AMA/INA 
AMA_INA_pumping = original_compiled_df.groupby(['AMA INA', 'YEAR'])['AF Pumped'].agg(['sum', 'mean'])
AMA_INA_pumping = AMA_INA_pumping.reset_index()
AMA_INA_pumping = AMA_INA_pumping.rename(columns={'YEAR': 'year', 'sum': 'AF_sum', 'mean': 'AF_mean'})
AMA_INA_pumping
```

```python
AMA_INA_pumping['AMA INA'].unique()
```

```python
# renaming AMA-INAs to have consistancy 
rename_dict = {'DOUGLAS INA': 'DOUGLAS AMA', 'HARQUAHALA VALLEY INA': 'HARQUAHALA INA',
              'JOSEPH CITY INA': 'JOSEPH CITY INA', 'PHOENIX AMA': 'PHOENIX AMA',
              'PINAL AMA': 'PINAL AMA', 'PRESCOTT AMA':'PRESCOTT AMA', 
              'SANTA CRUZ AMA': 'SANTA CRUZ AMA', 'TUCSON AMA': 'TUCSON AMA'}
AMA_INA_pumping['AMA INA'] = AMA_INA_pumping['AMA INA'].map(rename_dict)
AMA_INA_pumping = AMA_INA_pumping.dropna()
AMA_INA_pumping
```

```python
# loading AMA_INA shapefile and calcuating area
AMA_INA_gdf = gpd.read_file('../../Data_main/shapefiles/shapes/AMA_and_INA_AZ/AZ_AMA_INA.shp')
AMA_INA_gdf = AMA_INA_gdf.to_crs('EPSG:26712')
AMA_INA_gdf['area_mm2'] = AMA_INA_gdf['geometry'].area * 1000 * 1000 # area in mm2
AMA_INA_gdf = AMA_INA_gdf[['BASIN_NAME', 'area_mm2']]
AMA_INA_gdf
```

```python
# merging AMA INA pumping df and shapefile df to join area info
AMA_INA_pumping = AMA_INA_pumping.merge(AMA_INA_gdf, left_on='AMA INA', right_on='BASIN_NAME')
AMA_INA_pumping
```

```python
# saving the dataframe
AMA_INA_pumping.to_csv('../../Data_main/Pumping/Arizona/pumping_AZ_v2.csv', index=False)
```

```python
# selecting columsn of interest
columns_of_interst = ['long_nad83', 'lat_nad83', 'gmd', 'source', 'AF_USED_IRR_2000', 'AF_USED_IRR_2001', 'AF_USED_IRR_2002', 
                      'AF_USED_IRR_2003', 'AF_USED_IRR_2004', 'AF_USED_IRR_2005', 'AF_USED_IRR_2006', 'AF_USED_IRR_2007', 
                      'AF_USED_IRR_2008', 'AF_USED_IRR_2009', 'AF_USED_IRR_2010', 'AF_USED_IRR_2011', 'AF_USED_IRR_2012', 
                      'AF_USED_IRR_2013', 'AF_USED_IRR_2014', 'AF_USED_IRR_2015', 'AF_USED_IRR_2016', 'AF_USED_IRR_2017', 
                      'AF_USED_IRR_2018', 'AF_USED_IRR_2019', 'AF_USED_IRR_2020', 
                      'ACRES_2000', 'ACRES_2001', 'ACRES_2002', 'ACRES_2003', 'ACRES_2004', 'ACRES_2005', 'ACRES_2006', 
                      'ACRES_2007', 'ACRES_2008', 'ACRES_2009', 'ACRES_2010', 'ACRES_2011', 'ACRES_2012', 'ACRES_2013', 
                      'ACRES_2014', 'ACRES_2015', 'ACRES_2016', 'ACRES_2017', 'ACRES_2018', 'ACRES_2019', 'ACRES_2020']
pumping_KS_df = pumping_KS_df[columns_of_interst]
pumping_KS_df.head()
```

```python
# Using pandas melt to convert the wide format data to long format

af_columns = pumping_KS_df.filter(like='AF').columns
acres_columns = pumping_KS_df.filter(like='ACRES').columns

# for acre-ft
af_melted = pumping_KS_df.melt(id_vars=['lat_nad83', 'long_nad83', 'gmd', 'source'],
                               value_vars=list(af_columns),
                               var_name='var1', value_name='AF_pumped')

# for Acres
acres_melted = pumping_KS_df.melt(id_vars=['lat_nad83', 'long_nad83', 'gmd', 'source'],
                                 value_vars=list(acres_columns),
                                 var_name='var2', value_name='Acres')
```

```python
# saving both data as csv
af_melted.to_csv('../../Data_main/Pumping/Kansas/csv/pumping_KS_organized_2000_2020.csv', index=False)
acres_melted.to_csv('../../Data_main/Pumping/Kansas/csv/pumping_KS_acres_2000_2020.csv', index=False)
```
